```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
col = ['formation_energy_ev_natom','bandgap_energy_ev']
X = df.drop(['percent_atom_ga', 'number_of_total_atoms']+col,axis=1)
X['1']=1
X['p_al2'] = X['percent_atom_al']**2
X['p_in2'] = X['percent_atom_in']**2
X['l_v_1_a2'] = X['lattice_vector_1_ang']**2
X['l_v_2_a2'] = X['lattice_vector_2_ang']**2
X['l_v_3_a2'] = X['lattice_vector_3_ang']**2
y = df[col]
x_test = df_test.drop(['percent_atom_ga', 'number_of_total_atoms'],axis=1)
x_test['1']=1
x_test['p_al2'] = x_test['percent_atom_al']**2
x_test['p_in2'] = x_test['percent_atom_in']**2
x_test['l_v_1_a2'] = x_test['lattice_vector_1_ang']**2
x_test['l_v_2_a2'] = x_test['lattice_vector_2_ang']**2
x_test['l_v_3_a2'] = x_test['lattice_vector_3_ang']**2
```

```python
#following our data exploration we divide two spacegroups (12 and 194) each onto 2 different clusters
##X.spacegroup.loc[np.logical_and(X.spacegroup==12, X['lattice_vector_2_ang']<5)]=1012
##X.spacegroup.loc[np.logical_and(X.spacegroup==194, X['lattice_vector_2_ang']<5)]=1194

##X.spacegroup.unique()
```

```python
models = []
preds = []
for sg in X.spacegroup.unique():
    lr = LinearRegression()
    if sg != 1194:
        lr.fit(X.loc[X.spacegroup==sg],y.loc[X.spacegroup==sg])
        preds.append({'id': x_test.loc[x_test.spacegroup==sg].index.get_values(), 
                  'formation_energy_ev_natom' : lr.predict(x_test.loc[x_test.spacegroup==sg])[:,0],
                  'bandgap_energy_ev' : lr.predict(x_test.loc[x_test.spacegroup==sg])[:,1]})
    else:
        lr.fit(X.loc[X.spacegroup==1012].drop(['lattice_vector_3_ang','l_v_3_a2'], axis=1),y.loc[X.spacegroup==1012])
        preds.append({'id': x_test.loc[x_test.spacegroup==sg].index.get_values(), 
                      'formation_energy_ev_natom' : lr.predict(x_test.loc[x_test.spacegroup==sg].drop(['lattice_vector_3_ang','l_v_3_a2'], axis=1))[:,0],
                      'bandgap_energy_ev' : lr.predict(x_test.loc[x_test.spacegroup==sg].drop(['lattice_vector_3_ang','l_v_3_a2'], axis=1))[:,1]})
    models.append(lr)
```

```python
preds
```

```python
submit_df=0
for item in preds:
    if type(submit_df)==pd.core.frame.DataFrame:
        submit_df = submit_df.append(pd.DataFrame(item))
    else:
        submit_df = pd.DataFrame(item)
print(submit_df.info())
```

```python
submit_df = submit_df.sort_values('id').set_index('id')[['formation_energy_ev_natom', 'bandgap_energy_ev']]
submit_df.to_csv('submit1.csv')
submit_df
```
